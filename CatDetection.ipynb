```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
cat_img = cv2.imread('cat.jpg')
cat_img = cv2.resize(cat_img, (500,500))
```

```python
face_cascade = cv2.CascadeClassifier("haarcascades/haarcascade_frontalcatface_extended.xml")
```

```python
image = cv2.cvtColor(cat_img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

face = face_cascade.detectMultiScale(gray, 1.1, 15)

c_image = image.copy()
for (x,y,w,h) in face:
    cv2.rectangle(c_image, (x,y), ((x+w),(y+h)), (0,255,0), 3)
```

```python
plt.imshow(c_image)
plt.axis('off')
plt.show()
```
