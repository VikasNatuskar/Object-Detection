```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
p_image = cv2.imread('people.jpg')
body_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_fullbody.xml')

#img = cv2.cvtColor(p_image, cv2.COLOR_BGR2RGB)
#plt.figure(figsize=[10,10])
plt.imshow(p_image)
plt.show()
```

```python
import cv2
gray = cv2.cvtColor(p_image, cv2.COLOR_BGR2GRAY)
body = body_cascade.detectMultiScale(gray, 1.03, 3)
```

```python
simage=p_image.copy()
for (bx,by,bw,bh) in body:
    cv2.rectangle(simage,(bx, by),((bx + bw),(by + bh)),(0,255,0),2)
```
